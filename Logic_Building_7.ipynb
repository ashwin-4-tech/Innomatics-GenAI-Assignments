### Problem 1: Smart Light Controller

```python
class SmartLight:
    def __init__(self, name):
        self.name = name
        self.status = "OFF"

    def turn_on(self):
        self.status = "ON"
        print(f"{self.name} is {self.status}")

    def turn_off(self):
        self.status = "OFF"
        print(f"{self.name} is {self.status}")

    def display_status(self):
        print(f"{self.name} is {self.status}")



bedroom_light = SmartLight("Bedroom Light")
bedroom_light.display_status()
```

### Problem 2: Employee ID Card System

```python
class Employee:
    def __init__(self, name, employee_id, department):
        self.name = name
        self.employee_id = employee_id
        self.department = department

    def display_id_card(self):
        print("╔═══════════════════════════════════╗")
        print("║          EMPLOYEE ID CARD         ║")
        print("╠═══════════════════════════════════╣")
        print(f"║ Name: {self.name:<25}║")
        print(f"║ ID:   {self.employee_id:<25}║")
        print(f"║ Dept: {self.department:<25}║")
        print("╚═══════════════════════════════════╝")

employee_details = Employee(input("Name: "), input("ID: "), input("Department: "))
employee_details.display_id_card()
```

### Problem 3: Mobile Pbonebook (Contacts)

```python
class Contact:
    def __init__(self, name, phone_number):
        self.name = name
        self.phone_number = phone_number

    def display_contact_info(self):
        print("Contact Saved")
        print(f"Name: {self.name}")
        print(f"Phone: {self.phone_number}")

name_input = input("Name: ")
phone_number_input = ""
while True:
    phone_number_input = input("Phone: ")
    if phone_number_input.isdigit() and len(phone_number_input) == 10:
        break
    else:
        print("Error: Phone number must contain exactly 10 digits. Please try again.")

anita = Contact(name_input, phone_number_input)
anita.display_contact_info()
```

### Problem 4: Product Price Tag Generator

```python
class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price

    def print_price_tag(self):
        print(f"Product: {self.name.upper()}")
        print(f"Price: ₹{self.price}")

product_details = Product(name=input("Product Name: "), price=int(input("Price: ")))
product_details.print_price_tag()
```

### Problem 5: Movie Rating Display System

```python
class Movie:
    def __init__(self, name, rating):
        self.name = name
        self.rating = rating

    def display_movie_details(self):
        print("╔═══════════════════════════════════╗")
        print("║{:^35}║".format("LetterBOXD"))
        print("╠═══════════════════════════════════╣")
        print(f"║ Movie: {self.name.upper():<27}║")
        print(f"║ Rating: {f'{self.rating:.1f}':<22\}║")
        print("╚═══════════════════════════════════╝")


movie_name_input = input("Movie Name: ")
movie_rating_input = float(input("Rating: "))
film_details = Movie(movie_name_input, movie_rating_input)
film_details.display_movie_details()
```

### Problem 6: Delivery Address Manager

```python
class Delivery:
    def __init__(self, customer_name, address):
        self.customer_name = customer_name
        self.address = address

    def print_delivery_details(self):
        print("Delivery Details")
        print(f"Customer: {self.customer_name.upper()}")
        print(f"Pincode: {self.address}")

customer_name_input = input("Customer Name: ")
address_input = int(input("Pincode: "))

delivery = Delivery(customer_name_input, address_input)
delivery.print_delivery_details()
```
