Problem 1: Unique Words in a Sentence


---



```python
# Define the input sentence
sentence = "Python is easy and Python is powerful"

# Split the sentence into individual words based on spaces
words = sentence.split(" ")

# Convert the list of words into a set to automatically remove duplicate words
result = set(words)

# Print the count of unique words found in the sentence
print(f"Unique words count: {len(result)}")
# Print the set of unique words
print(f"Unique words: {result}")
```

Problem 2: Highest Salary from Employee Data


---



```python

employees_salary = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}


highest_salary = max(employees_salary.values())

highest_paid_employees = [name for name, salary in employees_salary.items() if salary == highest_salary]


print(f"Highest salary: {highest_paid_employees[0]} - {highest_salary}")
```

Problem 4: Count Products Above a Price Threshold


---



```python
prices = [450, 1200, 899, 1500, 300]
product_count = 0

for price in prices:
  if price > 1000:
    product_count += 1

print(f"Products above 1000 : {product_count}")
```

Problem 5: Calculate Attendance Percentage


---



```python
attendance = ["P", "P", "A", "P", "P"]

present_count = 0

for status in attendance:
  if status == "P":
    present_count += 1

percentage = (present_count / len(attendance)) * 100

print(f"Attendance percentage: {percentage:.0f}%")
```

Problem 7: Count Character Frequency


---



```python
string = "Python Programming"

char_frequency = {}
for char in string:
  if char != ' ':
    char_frequency[char.lower()] = char_frequency.get(char.lower(), 0) + 1

print(char_frequency)
```

Problem 10: Calculate Average Marks

---



```python
marks = [70, 80, 65, 75]

# Calculate the sum of the marks
total_marks = sum(marks)

# Calculate the number of marks
number_of_marks = len(marks)

# Calculate the average
average_marks = total_marks / number_of_marks

print(f"Average Marks: {average_marks}")
```
