Problem Statement 1: Employee Performance Bonus Eligibility


---



```python
employees = {
"Ravi": 92,
"Anita": 88,
"Kiran": 92,
"Suresh": 85
}

highest_score = 0
for score in employees.values():
    if score > highest_score:
        highest_score = score

top_performers = []
for name, score in employees.items():
    if score == highest_score:
        top_performers.append(name)

# Format the output as one line
output_string = f"Top Performers Eligible for Bonus: {', '.join(top_performers)} (Score: {highest_score})"

print(output_string)
```

Problem Statement 2: Search Query Keyword Analysis


---



```python
import string
from collections import Counter

input_str = "Buy mobile phone buy phone online"

# Convert to lowercase
lower_str = input_str.lower()

# Split into words
words = lower_str.split()

# Count frequencies
word_counts = Counter(words)

# Filter for keywords searched more than once
result = {word: count for word, count in word_counts.items() if count > 1}

print(result)
```

Problem Statement 3: Sensor Data Validation


---



```python
sensor_readings = [3, 4, 7, 8, 10, 12, 5]

# Use list comprehension with enumerate to filter valid readings
valid_readings = [(hour, reading) for hour, reading in enumerate(sensor_readings) if reading % 2 == 0]

print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

Problem Statement 4: Email Domain Usage Analysis


---



```python
emails = [
"ravi@gmail.com",
"anita@yahoo.com",
"kiran@gmail.com",
"suresh@gmail.com",
"meena@yahoo.com"
]

gmail = 0
yahoo = 0

for email in emails:
  if "@gmail.com" in email:
    gmail += 1
  else:
    yahoo += 1

total_emails = len(emails)

gmail_percentage = (gmail / total_emails) * 100
yahoo_percentage = (yahoo / total_emails) * 100

print(f"gmail.com: {gmail_percentage:.0f}%")
print(f"yahoo.com: {yahoo_percentage:.0f}%")
```

Problem Statement 5: Sales Spike Detection


---



```python
sales = [1200, 1500, 900, 2200, 1400, 3000]

# 1. Calculate Average
average_sales = sum(sales) / len(sales)

# 2. Calculate Threshold and force it to an integer
threshold = int(average_sales * 1.3)

threshold = round(threshold, -2)

# 3. Output results
for day, value in enumerate(sales, start=1):
    if value >= threshold:
        print(f"Day {day}: {value}")
```

Problem Statement 6: Duplicate User ID Detection


---



```python
from collections import Counter

user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

# Count the frequency of each user ID
id_counts = Counter(user_ids)

# Identify and display duplicate user IDs and their counts
for user_id, count in id_counts.items():
    if count > 1:
        print(f"{user_id} -> {count} times")
```
