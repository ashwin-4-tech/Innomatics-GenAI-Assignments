Problem 1: Mobile Recharge Validation System


---



```python
def recharge():
  valid_plans = [199, 299, 399, 599]

  while True:
    amount_input = input("Enter the recharge amount: ")
    recharge_amount = int(amount_input) # This line will raise ValueError if input is not an integer

    if recharge_amount not in valid_plans:
      print(f"Invalid plan. Please choose from: {', '.join(map(str, valid_plans))}. Please try again.")
    else:
      print(f"Proceeding to recharge with ₹{recharge_amount}.")
      return "Recharge successful!"


# Call the function to start the validation process
recharge()
```

Problem 2: Inventory Reorder Alert System


---



```python
def check_inventory(products_stock):
  reorder_threshold = 15
  reorder_alerts = []
  stock_ok = []

  for product, quantity in products_stock.items():
    if quantity < reorder_threshold:
      reorder_alerts.append(f"Reorder Alert for {product}: {quantity}")
    else:
      stock_ok.append(f"Stock OK for {product}: {quantity}")

  # Print reorder alerts first
  for alert in reorder_alerts:
    print(alert)
  print("--------------------")

  # Then print stock OK messages
  for ok_item in stock_ok:
    print(ok_item)

# Example usage:
inventory = {
    "Laptop": 20,
    "Mouse": 10,
    "Keyboard": 50,
    "Monitor": 12,
    "Webcam": 25
}

print("\n--- Inventory ---")
check_inventory(inventory)
```

Problem 3: Student Result Processing System


---






```python
def process_student_marks(marks):
  if not marks:
    print("No marks provided.")
    return

  total_marks = sum(marks)
  average_marks = total_marks / len(marks)

  print(f"Student Average Marks: {average_marks:.2f}")

  if average_marks >= 50:
    print("Result: Pass")
  else:
    print("Result: Fail")

# Example usage:
student1_marks = [70, 85, 60, 92, 45]
print("\n--- Student 1 Results ---")
process_student_marks(student1_marks)

student2_marks = [30, 40, 48, 55, 35]
print("\n--- Student 2 Results ---")
process_student_marks(student2_marks)

student3_marks = []
print("\n--- Student 3 Results ---")
process_student_marks(student3_marks)
```

Problem 4: Cab Fare Estimator with Retry Option


---



```python
def calculate_fare(distance):
    base_fare = 50
    rate_per_km = 12
    return base_fare + (distance * rate_per_km)

print("--- Fare Calculator ---")

while True:
    # Receive input strictly as an integer
    distance = int(input("\nEnter distance in km: "))

    # Calculate using the function
    total = calculate_fare(distance)

    # Display the final amount
    print(f"Total Fare: ₹{total}")

    # Retry logic
    retry = input("\nCalculate another fare? (y/n): ").strip().lower()
    if retry != "y":
        print("Program closed. Have a great day!")
        break
```

Problem 5: Employee Attendance Eligibility Checker


---



```python
def check_eligibility(attendance_list):
  if not attendance_list:
    print("Attendance list is empty. Cannot determine eligibility.")
    return "Invalid"

  total_days = len(attendance_list)
  present_days = attendance_list.count('P')

  attendance_percentage = (present_days / total_days) * 100

  print(f"Attendance Percentage: {attendance_percentage:.0f}%")

  if attendance_percentage >= 75:
    return "Eligible"
  else:
    return "Not Eligible"

# Example usage:
student1_attendance = ['P', 'P', 'A', 'P', 'P', 'P', 'A', 'P', 'P', 'P']
print("\n--- Employee 1 Eligibility ---")
print(check_eligibility(student1_attendance))

student2_attendance = ['P', 'A', 'A', 'P', 'A', 'P', 'A', 'A', 'P', 'A']
print("\n--- Employee 2 Eligibility ---")
print(check_eligibility(student2_attendance))

student3_attendance = []
print("\n--- Employee 3 Eligibility ---")
print(check_eligibility(student3_attendance))
```

Problem 6: Password Strength Checker

---



```python
def password_strength(password):
  min_length = 8
  has_digit = False
  has_special_char = False
  special_chars = "@#$"

  # Check length
  if len(password) < min_length:
    return "Weak: Password must be at least 8 characters long."

  # Check for digit and special character
  for char in password:
    if char.isdigit():
      has_digit = True
    elif char in special_chars:
      has_special_char = True

  if not has_digit:
    return "Weak: Password must contain at least one digit."
  if not has_special_char:
    return "Weak: Password must contain at least one special character (@#$)."

  return "Strong Password."

# Get password input from the user with retry logic
while True:
  user_password = input("Enter your password: ")
  strength_result = password_strength(user_password)
  print(strength_result)

  if strength_result.startswith("Strong"):
    print("Password accepted!")
    break
  else:
    print("Please try again with a stronger password.")
```
