```python
# This cell analyzes a list of marks to count how many students passed or need to reappear.
# A student passes if their mark is 50 or above.
# It then prints the count of 'Passed Out' and 'Reappear' students.
marks = [45, 78, 90, 33, 60]

passed_out = 0
reappear = 0

for mark in marks:
  if mark >= 50:
    passed_out += 1
  else:
    reappear += 1

print(f"Passed Out = {passed_out}")
print(f"Reappear = {reappear}")
```

```python
# This cell analyzes a list of messages to find their lengths and flag long ones.


messages = ["Hi", "Welcome to the platform", "OK"]

flagged_messages = [] # This list will store messages that are flagged

for message in messages:
  print(len(message)) # Print the length of the current message

  if len(message) > 10: # Check if the message is longer than 10 characters
    flagged_messages.append(f"Flagged: {message}") # Add the flagged message to our list

# After checking all messages, print out all the collected flagged messages
for flagged_msg in flagged_messages:
  print(flagged_msg)
```

```python

logs = ["INFO", "ERROR", "WARNING", "ERROR"]

error_count = 0 # Initialize  counter for errors

for log_entry in logs: # Loop through logs list
  if log_entry == "ERROR":
    error_count += 1

print(f"Error Count - {error_count}")
```
